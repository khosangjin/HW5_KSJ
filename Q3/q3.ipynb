```python
import csv

if __name__ == "__main__":

    f = open('q3.csv')
    data = csv.reader(f)
    header = next(data)
    
    lines = ["1호선","2호선","3호선","4호선","5호선","6호선","7호선","8호선","9호선"]

    #0번 인덱스는 공백
    num = [0] * 10
    
    maxnum1 = 0
    maxnum2 = 0
    maxline1 = 0
    maxline2 = 0
    
    minnum1 = 99999999
    minnum2 = 99999999
    minline1 = 0
    minline2 = 0
    
    for row in data :
        
        #1~9 호선만
        if row[1] in lines :
            
            #호선 구분
            line = row[1].replace("호선","")          
            line = int(line)
            
            #이용객
            row[4] = int(row[4])
            row[5] = int(row[5])
            sumrow = row[4] + row[5]
            
            #호선 인덱스에 이용객 추가
            num[line] += sumrow
            
    f.close()    
    
    #최고 / 최저 구하기
    for i in range(1,10):
        
        if maxnum1 < num[i] and maxnum2 < num[i]:
            #최고
            maxnum2 = maxnum1
            maxline2 = maxline1
            
            maxnum1 = num[i]
            maxline1 = i
            
        if maxnum1 > num[i] and maxnum2 < num[i]:
            #최고2
            maxnum2 = num[i]
            maxline2 = i
            
        if minnum1 > num[i] and minnum2 > num[i]:
            #최저
            minnum2 = minnum1
            minline2 = minline1
            
            minnum1 = num[i]
            minline1 = i
            
        if minnum1 < num[i] and minnum2 > num[i]:
            #최고2
            minnum2 = num[i]
            minline2 = i
            
    print("*** Subway Report for Seoul on March 2023 ***\n")
    print("1st Busiest Line : Line{0:d} ({1:d})\n".format(maxline1,maxnum1))
    print("2nd Busiest Line : Line{0:d} ({1:d})\n".format(maxline2,maxnum2))
    print("1st Least Line : Line{0:d} ({1:d})\n".format(minline1,minnum1))
    print("2nd Least Line : Line{0:d} ({1:d})\n".format(minline2,minnum2))
```
