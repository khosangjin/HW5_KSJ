```python
import csv

if __name__ == "__main__":

    f = open('q4.csv')
    data = csv.reader(f)
    header = next(data)
    
    lines = ["1호선","2호선","3호선","4호선"]
    #1 /2 /3/ 4
    busiest_station = [""] * 5
    least_station = [""] * 5
    
    maxnum = [0] * 5
    minnum = [99999999] * 5
    
    maxtmp = [0] * 5
    mintmp = [99999999] *5
    
    
    for row in data :
        
        if row[1] in lines : 
            
            line = row[1].replace("호선","")
            
            line = int(line)
                       
            row[4] = int(row[4])
            row[5] = int(row[5])
            
            sumrow = row[4] + row[5]
            
            if sumrow > maxtmp[line] : 
                maxtmp[line] = sumrow
                maxnum[line] = sumrow
                busiest_station[line] = row[3]
                
            if sumrow < mintmp[line] : 
                mintmp[line] = sumrow
                minnum[line] = sumrow
                least_station[line] = row[3]
            
    f.close()    
    
    
    
    print("*** Subway Report for Seoul on March 2023 ***\n")
    print("Line1\n")
    print("Busiest Station:{0:s}({1:d})\n".format(busiest_station[1],maxnum[1]))
    print("Least used Station:{0:s}({1:d})\n\n".format(least_station[1],minnum[1]))
    print("Line2\n")
    print("Busiest Station:{0:s}({1:d})\n".format(busiest_station[2],maxnum[2]))
    print("Least used Station:{0:s}({1:d})\n\n".format(least_station[2],minnum[2]))
    print("Line3\n")
    print("Busiest Station:{0:s}({1:d})\n".format(busiest_station[3],maxnum[3]))
    print("Least used Station:{0:s}({1:d})\n\n".format(least_station[3],minnum[3]))
    print("Line4\n")
    print("Busiest Station:{0:s}({1:d})\n".format(busiest_station[4],maxnum[4]))
    print("Least used Station:{0:s}({1:d})\n".format(least_station[4],minnum[4]))
```
